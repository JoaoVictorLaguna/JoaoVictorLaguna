```python
nome = input('Digite o nome do vendedor: ')
janeiro = int(input('Digite o valor de vendas do mês 1: '))
fevereiro = int(input('Digite o valor de vendas do mês 2: '))
marco = int(input('Digite o valor de vendas do mês 3: '))
abril = int(input('Digite o valor de vendas do mês 4: '))
media = int(janeiro+fevereiro+marco+abril) / 4

print ('A media de vendas do vendedor',nome,'é',media)
```
